```python
#importing pandas
import pandas as pd
```

```python
import PyPDF2
```

```python
#importing python's regex
import re
```

```python
from PyPDF2 import PdfFileMerger,PdfFileReader
```

```python
object = PyPDF2.PdfFileReader("Part-1.pdf")
```

```python
NumPages = object.getNumPages()
```

```python
String = "Application No.202011013272 A"
```

```python
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    print("this is page " + str(i)) 
    Text = PageObj.extractText()
    search = re.search(String, Text)
    print(search)
```
